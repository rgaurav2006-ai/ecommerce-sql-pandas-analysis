# E-commerce Sales Analysis

Exploratory data analysis of e-commerce sales data using Pandas — covering revenue, top products, customer counts, monthly trends, and order status.

```python
# Import pandas
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("data/ecommerce_sales_dataset.csv")
```

```python
# Preview data
df.head()
```

```python
# Dataset structure
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# Total rows and columns
df.shape
```

```python
# Total revenue
df["Total_Price"].sum()
```

```python
# Total revenue in lakhs
round(df["Total_Price"].sum()/100000,2)
```

```python
# Unique customers
df["Customer_ID"].nunique()
```

```python
# Top 5 selling products
df.groupby("Product_Name")["Quantity"].sum().sort_values(ascending=False).head(5)
```

```python
# Revenue by product category
df.groupby("Product_Category")["Total_Price"].sum().sort_values(ascending=False)
```

```python
# Revenue by category in lakhs
round(df.groupby("Product_Category")["Total_Price"].sum()/100000,2)
```

```python
# Convert order date to datetime
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
# Monthly revenue trend
df.groupby(df["Order_Date"].dt.month)["Total_Price"].sum()
```

```python
# Cancelled orders
df[df["Order_Status"]=="Cancelled"].shape[0]
```

```python
# Orders by payment type
df.groupby("Payment_Type").size().sort_values(ascending=False)
```

```python
# Average order value
round(df["Total_Price"].mean(),2)
```
